```python
import pandas as pd

def get_excel_columns(file_path, sheet_name=0):
    """
    دریافت نام ستون‌های یک فایل اکسل.
    
    Parameters:
    - file_path: مسیر فایل اکسل
    - sheet_name: نام یا ایندکس شیت (پیش‌فرض 0 یعنی اولین شیت)
    
    Returns:
    - list: لیست نام ستون‌ها
    """
    try:
        df = pd.read_excel(file_path, sheet_name=sheet_name, nrows=0)  # فقط هدر را می‌خواند
        columns = df.columns.tolist()
        return columns
    except Exception as e:
        print(f"خطا در خواندن فایل: {e}")
        return []

# مثال استفاده
file = "Export.xlsx"
cols = get_excel_columns(file)
print("ستون‌ها:", cols)
```
